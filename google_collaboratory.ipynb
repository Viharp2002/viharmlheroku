Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection and Analysis

```python
# loading the diabetes datasets to a pandas Dataframe
diabetes_dataset = pd.read_csv('/content/diabetes.csv') 
```

```python
#Printing the first 5 rows of the datasets
diabetes_dataset.head()
```

```python
# Number of Rows and Columns
diabetes_dataset.shape
```

```python
# Getting the statistical data of this data
diabetes_dataset.describe()
```

```python
# Total Mumber of Diabetes Patients and Non-Diabetes Patients
diabetes_dataset['Outcome'].value_counts()
```

Here, '0' represents 'Non-Diabetes Patients'
       '1' represents 'Diabetes Patients'

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
# Separating the data and labels
x=diabetes_dataset.drop(columns='Outcome',axis=1)
# Here, axis=1 represents that a column is going to drop,and axis=0 means droping a row
y=diabetes_dataset['Outcome']
```

```python
print(x)
```

```python
print(y)
```

Data Standardization

```python
scaler= StandardScaler()
```

```python
scaler.fit(x)
```

```python
stanardized_data= scaler.transform(x)
```

```python
print(stanardized_data)
```

```python
x= stanardized_data
y= diabetes_dataset['Outcome']
```

```python
print(x)
print(y)
```

Train-Test Split

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

Training the Model

```python
classifier = svm.SVC(kernel='linear')
```

```python
#Training the Support Vector Machine Classifier
classifier.fit(x_train,y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy score on the training data
x_train_prediction= classifier.predict(x_train)
training_data_accuracy= accuracy_score(x_train_prediction, y_train)
```

```python
print('Accuracy score of the training data: ',training_data_accuracy)
```

```python
# accuracy score on the test data
x_test_prediction= classifier.predict(x_test)
testing_data_accuracy= accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy score of the testing data: ',testing_data_accuracy)
```
